```python
map = folium.Map(location=[37.77, -122.42])
incidents = folium.map.FeatureGroup()

for lat, long in zip(top_100.Y, top_100.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, long],
            radius=6,
            color="orange",
            fill=True,
            fill_color='red',
            fill_opacity= 0.5
        )
    )
map.add_child(incidents)
```

```python
map = folium.Map(location=[37.77, -122.42])
incidents = folium.map.FeatureGroup()

for lat, long in zip(top_100.Y, top_100.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, long],
            radius=6,
            color="orange",
            fill=True,
            fill_color='red',
            fill_opacity= 0.5
        )
    )

for lat, long, label in zip(top_100.Y, top_100.X, top_100.Descript):
    folium.Marker([lat, long], popup=label).add_to(map)

map.add_child(incidents)
```
